```python
What are the key properties of a Bernoulli distribution?

Ans : A Bernoulli distribution models a single trial with two possible outcomes, typically labeled as success (1) and failure (0). Its key properties include: binary outcomes, constant probability of success, independent trials, and complementary probability for failure. Additionally, the mean is equal to the probability of success (p), and the variance is p(1-p).
Here's a more detailed breakdown:
Binary Outcomes: A Bernoulli distribution is defined for a random variable that can only take two possible values, usually 0 and 1, representing failure and success, respectively.
Constant Probability: The probability of success, denoted by 'p', remains the same for each trial.
Independent Trials: The outcome of one trial does not affect the outcome of any other trial.
Complementary Probability: The probability of failure (q) is equal to 1 minus the probability of success (1-p).
Mean (Expected Value): The mean (or expected value) of a Bernoulli distribution is equal to p.
Variance: The variance of a Bernoulli distribution is equal to p(1-p)
```

```python
What is the binomial distribution, and how is it used in probability?

Ans: The binomial distribution is a discrete probability distribution that models the probability of achieving a specific number of successes in a fixed number of independent trials, where each trial has only two possible outcomes (success or failure).
 It's frequently used to analyze scenarios like coin flips, quality control, and survey responses, where the outcome of each trial is binary.
Key Characteristics of the Binomial Distribution:
Fixed Number of Trials: The number of trials (n) in the experiment is predetermined and constant.
Independent Trials: The outcome of one trial does not influence the outcome of any other trial.
Two Possible Outcomes: Each trial results in either success or failure.
Constant Probability of Success: The probability of success (p) remains the same for each trial.
How it's used in Probability:
The binomial distribution helps calculate the probability of obtaining a specific number of successes (x) in n trials, given the probability of success (p) for each trial.
Formula:
The probability of exactly x successes in n trials is calculated using the following formula:
Code

P(X = x) = (nCx) * p^x * (1-p)^(n-x)
Where:
P(X = x) is the probability of getting exactly x successes.
nCx is the number of combinations of n items taken x at a time (also written as n! / (x! * (n-x)!)).
p is the probability of success on a single trial.
1-p is the probability of failure.
x is the number of successes.
n is the number of trials.
Examples:
Coin Flipping:
You can use the binomial distribution to calculate the probability of getting exactly 3 heads in 5 coin flips, where success is defined as getting a head (and failure is getting a tail).
Quality Control:
In manufacturing, you can use it to determine the probability of finding a certain number of defective items in a batch, given the probability of a defective item.
Survey Responses:
If you're conducting a survey, you can use it to find the probability of getting a specific number of "yes" responses to a question.
In essence, the binomial distribution provides a powerful tool for analyzing and predicting outcomes in situations where there are a fixed number of independent trials with two possible results
```

```python
What is the application of Z statistics in hypothesis testing?

Ans: Z-statistics, calculated using a Z-test, are applied in hypothesis testing to determine if there's a significant difference between a sample's mean and a population's mean, or between the means of two independent samples. This is particularly useful when the population variance is known, or when dealing with large sample sizes (typically n ≥ 30).
Here's a more detailed breakdown:
1. Comparing Sample Mean to Population Mean:
One-sample Z-test:
This test is used when you want to compare the mean of a sample to a known population mean.
Example:
Determining if the average test score of a class differs significantly from the statewide average score.
2. Comparing Means of Two Samples:
Two-sample Z-test: This test helps determine if there's a significant difference between the means of two independent samples.
Example: Comparing the average income of residents in two different cities.
3. When to Use Z-tests:
Known Population Variance: When the population variance is known, a Z-test is appropriate.
Large Sample Sizes: When the sample size is large (n ≥ 30), even if the population variance is unknown, a Z-test can be used. In this case, the sample variance is used as an estimate of the population variance.
4. Key Steps in Hypothesis Testing with Z-tests:
State Hypotheses: Formulate the null and alternative hypotheses.
Choose Significance Level: Select a significance level (alpha, usually 0.05).
Calculate Z-statistic: Compute the Z-statistic using the appropriate formula for one-sample or two-sample Z-tests.
Determine Critical Value: Find the critical value(s) from the Z-table based on the chosen significance level and whether it's a one-tailed or two-tailed test.
Make Decision: Compare the calculated Z-statistic to the critical value(s).
If the Z-statistic falls within the rejection region (beyond the critical value(s)), reject the null hypothesis.
Otherwise, fail to reject the null hypothesis.
5. Z-test vs. T-test:
Z-tests are used when the population variance is known or when the sample size is large.
T-tests are used when the population variance is unknown and the sample size is small.
```

```python
What are point estimates and interval estimates in statistics?

Ans : In statistics, point estimates and interval estimates are two fundamental ways to estimate unknown population parameters using sample data.
Point estimates provide a single value as the best guess, while interval estimates provide a range of values within which the parameter is likely to fall.
Point Estimates:
A point estimate is a single value calculated from sample data that is used to estimate an unknown population parameter.
For example, the sample mean is a point estimate for the population mean, and the sample proportion is a point estimate for the population proportion.
Point estimates are straightforward but don't provide information about the uncertainty or variability in the estimation.
Interval Estimates:
An interval estimate provides a range of values, called a confidence interval, within which the population parameter is likely to lie, with a specified level of confidence.
For example, a 95% confidence interval for the population mean would be a range of values calculated from the sample data, such that we are 95% confident that the true population mean falls within that range.
Interval estimates account for the variability in the sample data and provide a measure of the uncertainty associated with the estimation.
Confidence intervals are a common type of interval estimate.
Interval estimates are preferred over point estimates when it is important to understand the potential range of values for the population parameter.
```

```python
What is the significance of confidence intervals in statistical analysis?

Ans: onfidence intervals in statistical analysis are crucial because they provide a range of plausible values for an unknown population parameter, offering a more informative measure of uncertainty than a single point estimate. They help researchers understand the reliability and precision of their estimates, guiding them in making data-driven decisions.
Here's a more detailed explanation:
1. Quantifying Uncertainty:
Confidence intervals express the uncertainty associated with an estimate derived from a sample.
Instead of just stating a single value (e.g., the average income of a group), a confidence interval provides a range within which the true population value is likely to fall.
For example, a 95% confidence interval of for the average height of adults in a city means that we are 95% confident that the true average height falls somewhere between 50 and 60 inches.
2. Reliability and Precision:
Wider confidence intervals indicate greater uncertainty and less precision in the estimate.
Conversely, narrower confidence intervals suggest a more precise estimate with less variability.
For example, if a survey has a wide confidence interval for the proportion of people who support a certain policy, it suggests that the estimate is less reliable than if the confidence interval were narrow.
3. Making Data-Driven Decisions:
Confidence intervals enable more informed decision-making by providing a range of plausible values rather than a single point.
For instance, in drug trials, confidence intervals around the effect size of a new medication can help determine if it's significantly better than an existing treatment.
If the confidence interval for the difference in effectiveness does not include zero, it suggests a statistically significant difference.
4. Complementing Significance Tests:
Confidence intervals offer a more complete picture than significance tests (p-values) alone.
While a p-value indicates whether a result is statistically significant, a confidence interval shows the magnitude and precision of the effect.
This is particularly helpful when dealing with borderline p-values, where the confidence interval can provide additional context.
5. Inferences about the Population:
Confidence intervals are essential for making inferences about a population based on a sample.
By calculating a confidence interval for a sample statistic, researchers can estimate the likely range of the corresponding population parameter.
For example, a confidence interval for the average income of a sample can be used to estimate the average income of the entire population from which the sample was drawn.
```

```python
What are the assumptions for applying the Central Limit Theorem?

Ans : The Central Limit Theorem (CLT) requires a few key assumptions to be met for the sampling distribution of sample means to approximate a normal distribution.
These include: random sampling, independence of samples, a sufficiently large sample size, and sometimes,
 that the sample size is less than 10% of the population when sampling without replacement.
Here's a more detailed explanation:
Random Sampling:
The sample must be drawn randomly from the population.
This ensures that each member of the population has an equal chance of being selected, which is crucial for reducing bias and making the sample representative of the population.
Independence of Samples:
The individual observations within the sample should be independent of each other.
This means that the selection of one data point should not influence the selection or value of any other data point in the sample.
Sufficiently Large Sample Size:
The sample size needs to be large enough for the CLT to be effective.
While a rule of thumb is often cited as n ≥ 30, the necessary sample size can vary depending on the shape of the population distribution.
For distributions that are heavily skewed or have outliers, a larger sample size might be needed.
Sampling Without Replacement:
If sampling is done without replacement, the sample size should not exceed 10% of the total population. This helps to maintain the independence of the samples.
These assumptions ensure that the sampling distribution of sample means will converge towards a normal distribution as the sample size increases,
regardless of the original population's distribution (as long as it has a finite variance).
```

```python
What is the concept of expected value in a probability distribution?

Ans : The expected value in a probability distribution represents the long-run average outcome of a random variable.
 its a weighted average of all possible values, where each value is weighted by its corresponding probability. Essentially, it tells you what you can "expect" to see, on average, if you repeat an experiment many times.
Here's a more detailed explanation:
Random Variable:
A random variable assigns a numerical value to each outcome of a random phenomenon. For example, the number of heads when flipping a coin three times is a random variable.
Probability Distribution:
This describes the likelihood of each possible outcome of the random variable. It can be represented as a table, graph, or formula.
Expected Value (E[X]):
For a discrete random variable, the expected value is calculated by summing the product of each possible outcome (x) and its probability (P(x)): E[X] = Σ [x * P(x)].
For a continuous random variable, it involves an integral: E[X] = ∫ x * f(x) dx, where f(x) is the probability density function.
Interpretation:
The expected value is not necessarily a value that can actually occur. Instead, it represents the average value you would expect to obtain if you repeated the random experiment many times.
In simpler terms: Imagine flipping a fair coin many times. The expected value of the number of heads you get in each flip is 0.5 (or 1/2), even though you can only ever get 0 or 1 in a single flip. This is because over many flips, the proportion of heads will approach 0.5.
Example: If you play a game where you win $10 with a probability of 0.2 and lose $2 with a probability of 0.8, the expected value of the game is (10 * 0.2) + (-2 * 0.8) = $0.40. This means that on average, you can expect to win $0.40 per game if you play it many times
```

```python
# Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Define the range for the discrete uniform distribution
# For example, a standard six-sided die roll (outcomes 1 to 6)
low_bound = 1
high_bound = 7  # Upper bound is exclusive in randint, so 7 means up to 6

# Generate the possible outcomes (x-values)
outcomes = np.arange(low_bound, high_bound)

# Calculate the Probability Mass Function (PMF) for each outcome
# The probability for each outcome in a discrete uniform distribution is 1 / (number of outcomes)
pmf_values = randint.pmf(outcomes, low=low_bound, high=high_bound)

# Plot the PMF
plt.figure(figsize=(8, 6))
plt.plot(outcomes, pmf_values, 'bo', ms=8, label='PMF values') # Plot points
plt.vlines(outcomes, 0, pmf_values, colors='b', lw=3, alpha=0.5, label='PMF bars') # Plot vertical lines
plt.title('Probability Mass Function (PMF) of a Discrete Uniform Distribution')
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.xticks(outcomes) # Ensure x-axis ticks align with outcomes
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

```python
# Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution

import math

def bernoulli_pdf(x, p):
    """
    Calculates the Probability Density Function (PDF) of a Bernoulli distribution.

    Args:
        x: The value at which to evaluate the PDF (0 or 1).
        p: The probability of success (a value between 0 and 1).

    Returns:
        The probability P(X = x), or None if x is not 0 or 1.
    """
    if x == 0:
        return 1 - p
    elif x == 1:
        return p
    else:
        return None

# Example usage:
probability_of_success = 0.7
x_value = 1
pdf_value = bernoulli_pdf(x_value, probability_of_success)

if pdf_value is not None:
    print(f"PDF at x = {x_value}: {pdf_value}")
else:
    print(f"Invalid input: x must be 0 or 1.")

x_value = 0
pdf_value = bernoulli_pdf(x_value, probability_of_success)

if pdf_value is not None:
    print(f"PDF at x = {x_value}: {pdf_value}")
else:
    print(f"Invalid input: x must be 0 or 1.")
```

```python
# Create a Poisson distribution and visualize it using Python

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import poisson

# Define the average rate of events (lambda, often denoted as mu in SciPy)
lambda_param = 5  # Example: average of 5 events per interval

# Generate a sample of data from a Poisson distribution
# using numpy.random.poisson for simulation
num_samples = 10000
poisson_data = np.random.poisson(lam=lambda_param, size=num_samples)

# Calculate the Probability Mass Function (PMF) for visualization
# using scipy.stats.poisson.pmf for theoretical probabilities
k_values = np.arange(0, int(max(poisson_data)) + 2) # Range of possible event counts
pmf_values = poisson.pmf(k_values, mu=lambda_param)

# Visualize the distribution
plt.figure(figsize=(10, 6))

# Histogram of the simulated data
sns.histplot(poisson_data, bins=np.arange(0, max(poisson_data) + 2) - 0.5,
             stat='density', kde=False, color='skyblue', label='Simulated Data')

# Plot the theoretical PMF
plt.plot(k_values, pmf_values, 'ro-', label='Theoretical PMF')

plt.title(f'Poisson Distribution (λ = {lambda_param})')
plt.xlabel('Number of Events')
plt.ylabel('Probability / Density')
plt.xticks(k_values)
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution

import numpy as np
import matplotlib.pyplot as plt

def cdf_discrete_uniform(x_values, lower_bound, upper_bound):
    """
    Calculates the CDF values for a discrete uniform distribution.

    Args:
        x_values (list or np.array): The values at which to calculate the CDF.
        lower_bound (int): The lower bound (inclusive) of the discrete uniform distribution.
        upper_bound (int): The upper bound (inclusive) of the discrete uniform distribution.

    Returns:
        np.array: The corresponding CDF values for each x_value.
    """
    if not (isinstance(lower_bound, int) and isinstance(upper_bound, int) and lower_bound <= upper_bound):
        raise ValueError("Lower and upper bounds must be integers and lower_bound <= upper_bound.")

    n = upper_bound - lower_bound + 1  # Number of possible outcomes
    cdf_values = []
    for x in x_values:
        if x < lower_bound:
            cdf_values.append(0)
        elif x >= upper_bound:
            cdf_values.append(1)
        else:
            # For discrete uniform, CDF(x) = (floor(x) - lower_bound + 1) / n
            # if x is within the range [lower_bound, upper_bound-1]
            cdf_values.append((np.floor(x) - lower_bound + 1) / n)
    return np.array(cdf_values)

def plot_cdf(x_values, cdf_values, lower_bound, upper_bound):
    """
    Plots the CDF of a discrete uniform distribution.

    Args:
        x_values (np.array): The x-values used for plotting.
        cdf_values (np.array): The corresponding CDF values.
        lower_bound (int): The lower bound of the discrete uniform distribution.
        upper_bound (int): The upper bound of the discrete uniform distribution.
    """
    plt.figure(figsize=(10, 6))
    plt.step(x_values, cdf_values, where='post', linestyle='-', marker='o', markersize=4, label='CDF')

    plt.title(f'Cumulative Distribution Function of Discrete Uniform Distribution ({lower_bound} to {upper_bound})')
    plt.xlabel('x')
    plt.ylabel('F(x) = P(X <= x)')
    plt.grid(True)
    plt.xticks(np.arange(lower_bound - 1, upper_bound + 2, 1)) # Adjust x-ticks for clarity
    plt.yticks(np.linspace(0, 1, 11)) # Show y-ticks from 0 to 1
    plt.ylim(-0.05, 1.05) # Add small buffer to y-axis limits
    plt.legend()
    plt.show()

# Example Usage:
if __name__ == "__main__":
    lower_bound = 1
    upper_bound = 6

    # Generate a range of x-values for plotting, including points outside the bounds
    x_plot = np.arange(lower_bound - 2, upper_bound + 3, 0.1)

    # Calculate CDF values
    cdf_vals = cdf_discrete_uniform(x_plot, lower_bound, upper_bound)

    # Plot the CDF
    plot_cdf(x_plot, cdf_vals, lower_bound, upper_bound)
```

```python
 # Write a Python function to calculate Z-scores from a dataset and plot them

 import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore

def calculate_and_plot_zscores(data):
    """
    Calculates Z-scores for a given dataset and plots their distribution.

    Args:
        data (list or np.array): The input dataset for which to calculate Z-scores.
    """
    if not isinstance(data, (list, np.ndarray)):
        raise TypeError("Input 'data' must be a list or a NumPy array.")
    if len(data) < 2:
        raise ValueError("Input 'data' must contain at least two elements to calculate Z-scores.")

    # Calculate Z-scores
    z_scores = zscore(data)

    # Plot the Z-score distribution
    plt.figure(figsize=(8, 6))
    plt.hist(z_scores, bins=20, edgecolor='black', alpha=0.7)
    plt.title('Distribution of Z-Scores')
    plt.xlabel('Z-Score')
    plt.ylabel('Frequency')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.axvline(0, color='red', linestyle='--', label='Mean (Z=0)') # Mark the mean (Z=0)
    plt.legend()
    plt.show()

    print("Calculated Z-scores:")
    print(z_scores)

# Example usage:
if __name__ == "__main__":
    # Create a sample dataset
    sample_data = [10, 12, 15, 18, 20, 22, 25, 30, 35, 40, 100]

    # Call the function to calculate and plot Z-scores
    calculate_and_plot_zscores(sample_data)

    # Example with a different dataset
    another_data = np.random.normal(loc=50, scale=10, size=100)
    calculate_and_plot_zscores(another_data)
```

```python
## Generate random variables and calculate their corresponding probabilities using the binomial distribution

import math

def binomial_probability(n, k, p):
  """
  Calculates the binomial probability.

  Args:
    n: The number of trials.
    k: The number of successes.
    p: The probability of success in a single trial.

  Returns:
    The binomial probability P(X=k).
  """
  combinations = math.comb(n, k) # or math.factorial(n) // (math.factorial(k) * math.factorial(n - k))
  probability = combinations * (p**k) * ((1-p)**(n-k))
  return probability

# Example usage:
n_trials = 5
num_successes = 3
probability_of_success = 0.5

probability = binomial_probability(n_trials, num_successes, probability_of_success)
print(f"The probability of {num_successes} successes in {n_trials} trials is: {probability}")
```

```python
# Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution

import numpy as np
from scipy.stats import norm

def calculate_z_score(data_point, mean, std_dev):
    """
    Calculates the Z-score for a given data point.

    Args:
        data_point (float): The individual data point.
        mean (float): The mean of the dataset.
        std_dev (float): The standard deviation of the dataset.

    Returns:
        float: The calculated Z-score.
    """
    z_score = (data_point - mean) / std_dev
    return z_score

def compare_to_standard_normal(z_score):
    """
    Compares the Z-score to a standard normal distribution.

    Args:
        z_score (float): The calculated Z-score.

    Returns:
        tuple: A tuple containing:
            - float: The cumulative probability (P(X <= z_score)).
            - float: The probability of observing a value more extreme than the data point (two-tailed).
    """
    cumulative_probability = norm.cdf(z_score)

    # For a two-tailed comparison, consider the absolute Z-score
    probability_more_extreme = 2 * (1 - norm.cdf(abs(z_score)))

    return cumulative_probability, probability_more_extreme

# Example Usage
if __name__ == "__main__":
    # Sample data
    data_point_value = 70
    dataset_mean = 78
    dataset_std_dev = 25

    # 1. Calculate the Z-score
    z = calculate_z_score(data_point_value, dataset_mean, dataset_std_dev)
    print(f"The Z-score for {data_point_value} is: {z:.4f}")

    # 2. Compare to the standard normal distribution
    cumulative_prob, two_tailed_prob = compare_to_standard_normal(z)
    print(f"Cumulative probability (P(X <= {data_point_value})) is: {cumulative_prob:.4f}")
    print(f"Probability of observing a value more extreme than {data_point_value} (two-tailed) is: {two_tailed_prob:.4f}")

    # Interpretation
    if z > 0:
        print(f"The data point is {abs(z):.2f} standard deviations above the mean.")
    elif z < 0:
        print(f"The data point is {abs(z):.2f} standard deviations below the mean.")
    else:
        print("The data point is equal to the mean.")
```

```python
## implement hypothesis testing using Z-statistics for a sample dataset

import numpy as np
from scipy.stats import norm

def one_sample_z_test(sample, pop_mean, pop_sd, alpha=0.05):
    """
    Performs a one-sample Z-test.

    Args:
        sample (list or numpy array): The sample data.
        pop_mean (float): The known population mean.
        pop_sd (float): The known population standard deviation.
        alpha (float): The significance level (default is 0.05).

    Returns:
        tuple: A tuple containing the Z-statistic, p-value, and a boolean indicating
               whether the null hypothesis should be rejected.
    """
    n = len(sample)
    sample_mean = np.mean(sample)
    z_statistic = (sample_mean - pop_mean) / (pop_sd / np.sqrt(n))
    p_value = 2 * (1 - norm.cdf(abs(z_statistic)))  # Two-tailed test
    reject_null = p_value < alpha

    return z_statistic, p_value, reject_null

# Example usage:
sample_data = [25, 28, 22, 27, 30, 24, 26]
population_mean = 25
population_std = 5
alpha_level = 0.05

z_score, p_val, reject = one_sample_z_test(sample_data, population_mean, population_std, alpha_level)

print(f"Z-statistic: {z_score:.2f}")
print(f"P-value: {p_val:.3f}")
print(f"Reject null hypothesis: {reject}")
```

```python
## Create a confidence interval for a dataset using Python and interpret the result

import numpy as np
from scipy import stats

# 1. Generate or load your dataset
data = np.array([12, 15, 14, 10, 13, 17, 14, 15, 16, 14])

# 2. Define the confidence level (e.g., 95%)
confidence_level = 0.95

# 3. Calculate sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # ddof=1 for sample standard deviation
sample_size = len(data)

# 4. Calculate the standard error of the mean (SEM)
sem = sample_std / np.sqrt(sample_size)

# 5. Calculate the confidence interval using the t-distribution
# (appropriate for small sample sizes or unknown population standard deviation)
degrees_freedom = sample_size - 1
confidence_interval = stats.t.interval(confidence_level, degrees_freedom, loc=sample_mean, scale=sem)

# 6. Print the results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Confidence Level: {confidence_level * 100:.0f}%")
print(f"Confidence Interval: ({confidence_interval[0]:.2f}, {confidence_interval[1]:.2f})")
```

```python
## Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean

import numpy as np
import scipy.stats as st

# 1. Generate data from a normal distribution
np.random.seed(42)  # for reproducibility
sample_size = 30
mean = 50
std_dev = 10
data = np.random.normal(mean, std_dev, sample_size)

# 2. Calculate the confidence interval
confidence_level = 0.95
degrees_of_freedom = sample_size - 1
sample_mean = np.mean(data)
standard_error = st.sem(data)
confidence_interval = st.t.interval(confidence_level, degrees_of_freedom, sample_mean, standard_error)

# 3. Interpret the results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval for the mean: {confidence_interval}")
print(f"The confidence interval ( {confidence_interval[0]:.2f}, {confidence_interval[1]:.2f} ) provides a range within which we are {int(confidence_level*100)}% confident that the true population mean lies.")
```

```python
## Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_normal_pdf(mean, std_dev):
    """
    Calculates and visualizes the PDF of a normal distribution.

    Args:
        mean (float): The mean (mu) of the normal distribution.
        std_dev (float): The standard deviation (sigma) of the normal distribution.
    """

    # Generate x-values for the plot, typically a range around the mean
    # spanning a few standard deviations to capture most of the distribution.
    x = np.linspace(mean - 4 * std_dev, mean + 4 * std_dev, 500)

    # Calculate the PDF values for the given x-values using scipy.stats.norm
    pdf_values = norm.pdf(x, loc=mean, scale=std_dev)

    # Plot the PDF
    plt.figure(figsize=(8, 6))
    plt.plot(x, pdf_values, label=f'Normal Distribution (μ={mean}, σ={std_dev})', color='blue')
    plt.title('Probability Density Function (PDF) of a Normal Distribution')
    plt.xlabel('X-value')
    plt.ylabel('Probability Density')
    plt.grid(True)
    plt.legend()
    plt.show()

# Example usage:
# Define the parameters of the normal distribution
mu = 0  # Mean
sigma = 1  # Standard deviation

# Call the function to plot the PDF
plot_normal_pdf(mu, sigma)

# Another example with different parameters
mu_2 = 10
sigma_2 = 2
plot_normal_pdf(mu_2, sigma_2)
```

```python
## Simulate a random variable using a continuous uniform distribution and calculate its expected value

import random
import numpy as np

def simulate_uniform(a, b, n):
    """
    Simulates n random variables from a uniform distribution [a, b]
    and calculates the expected value.
    """
    # Generate random numbers
    random_numbers = [a + (b - a) * random.random() for _ in range(n)]

    # Calculate the expected value
    expected_value = (a + b) / 2

    # Calculate the sample mean
    sample_mean = np.mean(random_numbers)

    return random_numbers, expected_value, sample_mean

# Example usage:
a = 2
b = 10
n = 10000

random_numbers, expected_value, sample_mean = simulate_uniform(a, b, n)

print(f"Expected value: {expected_value}")
print(f"Sample mean: {sample_mean}")
print(f"Difference: {abs(expected_value - sample_mean)}")

# Verification:  The difference between the expected value and the sample mean should be small, especially for larger n.
```

```python
## Write a Python program to compare the standard deviations of two datasets and visualize the difference

import numpy as np
import matplotlib.pyplot as plt

def compare_and_visualize_std(dataset1, dataset2, name1="Dataset 1", name2="Dataset 2"):
    """
    Calculates the standard deviations of two datasets and visualizes the comparison.

    Args:
        dataset1 (list or np.array): The first dataset.
        dataset2 (list or np.array): The second dataset.
        name1 (str): The name for the first dataset (for visualization).
        name2 (str): The name for the second dataset (for visualization).
    """

    # Calculate standard deviations
    std_dev1 = np.std(dataset1)
    std_dev2 = np.std(dataset2)

    print(f"Standard Deviation of {name1}: {std_dev1:.4f}")
    print(f"Standard Deviation of {name2}: {std_dev2:.4f}")

    # Visualize the difference
    labels = [name1, name2]
    std_devs = [std_dev1, std_dev2]

    plt.figure(figsize=(8, 6))
    plt.bar(labels, std_devs, color=['skyblue', 'lightcoral'])
    plt.ylabel("Standard Deviation")
    plt.title("Comparison of Standard Deviations")
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

# Example Usage:
if __name__ == "__main__":
    # Create sample datasets
    data_a = [10, 12, 11, 13, 10, 14, 9, 11, 12, 10]
    data_b = [5, 15, 8, 18, 6, 17, 7, 16, 9, 14]

    compare_and_visualize_std(data_a, data_b, "Data A", "Data B")

    print("\n--- Another Example ---")
    data_c = np.random.normal(loc=0, scale=1, size=100) # Normally distributed data with std 1
    data_d = np.random.normal(loc=0, scale=2, size=100) # Normally distributed data with std 2

    compare_and_visualize_std(data_c, data_d, "Normal Data (Std 1)", "Normal Data (Std 2)")
```
